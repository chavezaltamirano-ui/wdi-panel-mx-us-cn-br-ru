Fase 2
Importación.

```python
import requests
import pandas as pd
import os
```

Fase 3
Definición de países.

```python
countries = [
    "MX",
    "US",
    "CN",
    "BR",
    "RU"
]
```

FASE 4. Descargar el catálogo de indicadores

```python
url = "https://api.worldbank.org/v2/indicator?format=json&per_page=30000"

r = requests.get(url)

catalogo_json = r.json()
```

FASE 5. Convertir el catálogo a DataFrame

```python
catalogo = pd.json_normalize(
    catalogo_json[1]
)
```

FASE 6. Ver cuántos indicadores existen

```python
print(
    "Número de indicadores:",
    len(catalogo)
)
```

FASE 7. Guardar el catálogo

```python
catalogo.to_csv(
    "catalogo_indicadores.csv",
    index=False
)
```

FASE 8. Crear la lista de códigos

```python
codigos = catalogo["id"].tolist()
```

FASE 9. Verificar algunos códigos

```python
print(codigos[:10])
```

FASE 10. Preparar contenedor de datos

```python
todos_los_datos = []
```

FASE 11. Descargar los indicadores

```python
for n, indicador in enumerate(codigos[:400], start=1):

    print(f"Indicador {n}: {indicador}")

    for pais in countries:

        try:

            url = (
                f"https://api.worldbank.org/v2/"
                f"country/{pais}/"
                f"indicator/{indicador}"
                f"?format=json&per_page=1000"
            )

            r = requests.get(url)

            datos = r.json()

            if len(datos) > 1:

                df = pd.DataFrame(datos[1])

                df["pais"] = pais
                df["indicador"] = indicador

                todos_los_datos.append(df)

        except:
            pass
```

FASE 12. Unir todas las descargas

```python
df_total = pd.concat(
    todos_los_datos,
    ignore_index=True
)
```

FASE 13. Verificar dimensiones

```python
print(type(todos_los_datos))
print(len(todos_los_datos))
```

```python
df_total = pd.concat(
    todos_los_datos,
    ignore_index=True
)

print(df_total.shape)
```

FASE 14. Contar indicadores efectivos

```python
print(
    "Indicadores efectivos:",
    df_total["indicador"].nunique()
)
```

FASE 15. Verificar registros

```python
print(
    "Registros actuales:",
    len(df_total)
)
```

FASE 16. Convertir la fecha a numérica

```python
df_total["date"] = pd.to_numeric(
    df_total["date"],
    errors="coerce"
)
```

FASE 17. Filtrar 1990-2025

```python
df_total = df_total[
    (df_total["date"] >= 1990)
    &
    (df_total["date"] <= 2025)
]
```

FASE 18. Verificar registros después del filtro

```python
print(
    "Registros después del filtro:",
    len(df_total)
)
```

FASE 19. Contar faltantes

```python
print(
    "Valores faltantes:",
    df_total["value"].isna().sum()
)
```

FASE 20. Limpiar faltantes

```python
df_total = df_total.dropna(
    subset=["value"]
)
```

FASE 21. Contar registros válidos

```python
print(
    "Registros válidos:",
    len(df_total)
)
```

FASE 22. Guardar el dataset en CSV

```python
df_total.to_csv(
    "WDI_completo_panel.csv",
    index=False
)
```

FASE 23. Guardar en formato Parquet


```python
df_total.to_parquet(
    "WDI_completo_panel.parquet",
    index=False
)
```
